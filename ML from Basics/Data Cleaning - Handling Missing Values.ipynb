```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
type(null_var)
```

```python
drop_cols = null_var[null_var>17].keys()
drop_cols
```

```python
df_drop_cols = df.drop(columns = drop_cols)
df_drop_cols.shape
```

```python
sns.heatmap(df_drop_cols.isnull())
```

```python
df_drop_rows = df_drop_cols.dropna()
df_drop_rows.shape
```

```python
sns.heatmap(df_drop_rows.isnull())
```
