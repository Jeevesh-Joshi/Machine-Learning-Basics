## Pandas Series DataStructure

```python
import pandas as pd
pd.__version__
```

```python
list_s = [1,2,-3,"Hello","World"]
list_s
```

```python
series1 = pd.Series(list_s) # creating series data structure in pandas
print(series1)
```

```python
type(series1)
```

```python
# or,
series2 = pd.Series([1,2,3,4])
print(series2)
```

```python
empty_s = pd.Series([]) #Empty series
print(empty_s)
```

```python
series3 = pd.Series([1,2,3,4], index = ['a','b','c','d']) # Giving user defined index
print(series3)
```

```python
series4 = pd.Series([1,2,3,4], dtype = float) #Changing default dtype
print(series4)
```

```python
series5 = pd.Series([1,2,3,4], name = 'values') #Giving name to data values column
print(series5)
```

```python
# Creating series with single scalar value
scalar_s = pd.Series(0.5)
print(scalar_s)
```

```python
scalar_s = pd.Series(0.5, index=[1,2,3]) # for multiple values mention index
print(scalar_s)
```

```python
# creating series via dictionary
dict_s = pd.Series({'a':1,'b':2})
print(dict_s)
```

## Handling Missing values in Pandas

1. Pandas treat missing values by showing NaN where ever it founds value missing.
2. If there are values like 'not available' or 'no value' in your datasets, then Pandas won't show it as NaN and will show it as it is. Inorder to make Pandas treat them as NaN we use 'na_values' parameter, where we mention what all values to consider as NaN. Simply write na_values = ['not available', 'no values'] and pandas will show NaN in place of them.
3. To make Pandas consider specific column values as NaN use dictionary i.e na_values = {'Companies':'no values', 'Growth':'not available'}, specifying columnn nameand value as key,value pair.
4. In order not to show NaN, for all those strings that Pandas consider as NaN use 'keep_default_na' parameter i.e keep_default_na = False to not consider and True to consider.
5. If you are sure that the datasets does not conatin any missing vlaues then you can boost your performance by using 'na_filter'. By doing show Pandas now will not look for missing values, i,e use na_filter = False to boost your model performance.

#### 2. dropna()
It removes all the rows containing NaN values

```python
df
```

```python
df.dropna()
```

```python
1# if you want to remove all the columns containing NaN values provide axis as 1
df.dropna(axis=1)
```

2# how parameter is by default -> how = 'any' ;which means it will remove the rows/column (depending upon axis specified)containing NaN, whereas how = 'all' will only delete only if all the rows or cols are Nan.

```python
5# inplace parameter when set True will update the dataframe itself and does not creates a new one.
df.dropna(inplace=True)
df
```

#### 3. fillna()
It fills the NaN values with given values (input)

```python
df = pd.read_csv('D:\ML\ML from Basics\Datasets\Missing Values.csv')
df
```

```python
df.fillna(0) # NaN is replaced with the value specified
```

```python
# To fill specific column with specific values use dictionary
df.fillna({'Industry':'None','Revenue':0,'Growth':0})
```

```python
1# method parameter will fill NaN values with nearby values (depending on to what it is set).
df.fillna(method = 'ffill') # 'ffill'  and 'pad' takes previous value
```

```python
df.fillna(method = 'bfill') # 'bfill' takes next value
```

```python
2# axis tell from where to fill values in case of method i.e to fetch form column or from row
# when using asix always specify either 'method' or 'value'
df.fillna(axis=0, method='ffill') 
```

```python
df.fillna(axis=1, method='ffill') 
```

```python
df.fillna(axis=1, method='bfill') 
```

```python
3# limit parameter tells us how many values to fill in each column
# i,e if limit = x, then only first x values in each column will be filled with the value specified
df.fillna(0, limit=1) 
```

```python
df.fillna(method='ffill', limit=2) 
```

```python
4# inplace parameter when set True will update the current dataframe
df.fillna(5, inplace=True) 
df
```
