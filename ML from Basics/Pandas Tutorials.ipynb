## Handling Missing values in Pandas

1. Pandas treat missing values by showing NaN where ever it founds value missing.
2. If there are values like 'not available' or 'no value' in your datasets, then Pandas won't show it as NaN and will show it as it is. Inorder to make Pandas treat them as NaN we use 'na_values' parameter, where we mention what all values to consider as NaN. Simply write na_values = ['not available', 'no values'] and pandas will show NaN in place of them.
3. To make Pandas consider specific column values as NaN use dictionary i.e na_values = {'Companies':'no values', 'Growth':'not available'}, specifying columnn nameand value as key,value pair.
4. In order not to show NaN, for all those strings that Pandas consider as NaN use 'keep_default_na' parameter i.e set  keep_default_na = False to not consider and True to consider.
5. If you are sure that the datasets does not conatin any missing vlaues then you can boost your performance by using 'na_filter'. By doing show Pandas now will not look for missing values, i.e set na_filter = False to boost your model performance.

#### 9. merge()
It connects columns or indices in Dataframe based on one or more keys.

```python
df1 = pd.DataFrame({'ID':[1,2,3,4], 'Class':[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4], 'Section':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2,on='ID') #'on' will take that col name which is common in both dataframes
```

1. default join in merge is 'inner'. 
2. The length of entries in each col should be equal and also equal to the length of parameter specified to 'on'.

```python
df3 = pd.DataFrame({'ID':[1,2,3,5], 'Section':['A','B','C','D']})
pd.merge(df1,df3,on='ID') # you can see only 3 rows are being same as all ID's are not same in df3 and df1
```

```python
pd.merge(df1,df3,on='ID',how='inner') #default
```

```python
pd.merge(df1,df3,on='ID',how='left') #left outer join
```

```python
pd.merge(df1,df3,on='ID',how='right') #right outer join
```

```python
pd.merge(df1,df3,on='ID',how='outer') #full outer join
```

```python
# to check which value belongs to which df, set indicator = True
pd.merge(df1,df3,on='ID',how='outer',indicator=True)
```

```python
# if all the items are different then set 'left_index= True', 'right_index=True'.
df4 = pd.DataFrame({'ID':[5,6,7,8], 'Section':['A','B','C','D']})
pd.merge(df1,df4, left_index= True,right_index=True)
```

```python
# if cols are overlapping, it will show them like this, coz of 'suffixes' parameter which default x,y..
df5 = pd.DataFrame({'ID':[5,6,7,8], 'Section':['A','B','C','D']})
pd.merge(df4,df5,on='ID')
```

```python
pd.merge(df4,df5,on='ID', suffixes = ('_higher','_lower'))
```
