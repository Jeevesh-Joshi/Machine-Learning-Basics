```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.histplot(numeric_df[var],bins=20,kde=True,color='b') #orginal
    sns.histplot(median_df[var],bins=20,kde=True, color='y')  #mean  
```

# 4. Creating Pipeline 

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
train = pd.read_csv(r'Datasets\Kaggle\train.csv')
test = pd.read_csv(r'Datasets\Kaggle\test.csv')
train.shape
```

```python
X_train = train.drop(columns='SalePrice')
y_train = train['SalePrice']
X_test = test.copy()
```

```python
isnull_sum = X_train.isnull().sum()
isnull_sum
```

```python
# finding numerical values containing null values
num_vars = X_train.select_dtypes(include=['int64','float64']).columns
num_vars_miss = [var for var in num_vars if isnull_sum[var]>0]
num_vars_miss
```

```python
# finding categorical values containing null values
ctg_vars = X_train.select_dtypes(include=['object']).columns
ctg_vars_miss = [var for var in ctg_vars if isnull_sum[var]>0]
ctg_vars_miss
```

```python
num_vars_mean = ['LotFrontage']
num_vars_median = ['MasVnrArea', 'GarageYrBlt']
ctg_vars_mode = ['Alley','MasVnrType','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical','FireplaceQu']
ctg_vars_const = ['GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature']
```

```python
# Creating Pipeline for each strategy
num_vars_mean_imputer = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean'))])
num_vars_median_imputer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median'))])
ctg_vars_mode_imputer = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent'))])
ctg_vars_const_imputer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant',fill_value='missing'))])
```

```python
# transformers takes = ['name', 'imputer type', 'list of var in which you want to apply that imputer']
preprocessor = ColumnTransformer(transformers=[('mean_imputer', num_vars_mean_imputer, num_vars_mean),
                               ('median_imputer', num_vars_median_imputer, num_vars_median),
                               ('mode_imputer', ctg_vars_mode_imputer, ctg_vars_mode),
                               ('const_imputer', ctg_vars_const_imputer, ctg_vars_const)])
```

```python
# feeding values to our training data
preprocessor.fit(X_train)
```

```python
preprocessor.transform #additional details
```

```python
# to see the value which is used by the imputer
preprocessor.named_transformers_['mean_imputer'].named_steps['imputer'].statistics_
```

```python
X_train['LotFrontage'].mean()
```

```python
preprocessor.named_transformers_['mode_imputer'].named_steps['imputer'].statistics_
```

```python
# updating training set with the imputed values
X_train_clean = preprocessor.transform(X_train) 
X_test_clean = preprocessor.transform(X_test) 
```

```python
X_train_clean
```

```python
X_train_clean.shape
```

transformers has a parameter as 'remainder' whose default value is 'drop', which when used will drop all the columns that does not contain any null values from the dataset. To avoid this set 'remainder' as 'passthrough'.

```python
preprocessor.transformers_
```

```python
X_train_clean_were_missing = pd.DataFrame(X_train_clean, columns=num_vars_mean+num_vars_median+ctg_vars_mode+ctg_vars_const)
X_train_clean_were_missing.head()
```

```python
X_train_clean_were_missing.isnull().sum().sum()
```
